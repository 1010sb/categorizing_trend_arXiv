```python
import dask.bag as db
import datetime
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
```

```python
df1 = pd.read_csv('arxiv_1986_2022.csv', low_memory=False)
df1.head()
```

```python
# Let us check how the research papers are evolving over time from 1986 to 2022
# Create a countplot of the year column
sns.countplot(x='year', data=df1)

# Add labels and show the plot
plt.xlabel('Publication Year')
plt.ylabel('Papers Count')
plt.title("(1986 - 2022)", fontsize=10)
plt.suptitle('Trends in Scientific Publication by Year', fontsize=14)
plt.xticks(rotation=90)
plt.savefig('Trends in Scientific Publication by Year (1986 - 2022).jpg',dpi=200,bbox_inches='tight')
plt.show()
```

```python
# Group the data by the 'authors' column and count the number of papers for each author
author_counts = df1.groupby('authors')['id'].count().sort_values(ascending=False)

# Take the top 20 authors with the most papers
top_authors = author_counts.head(15)

# Plot the results using a bar chart with seaborn
sns.barplot(y=top_authors.index, x=top_authors.values, orient = 'h')
plt.xlabel('Number of Papers')
plt.ylabel('Authors')
plt.title('(1986 - 2022)', fontsize=10)
plt.suptitle('Prominent Authors on arXiv', fontsize=16)
plt.xticks(rotation=90)
plt.savefig('Prominent Authors on arXiv (1986 - 2022).jpg',dpi=200,bbox_inches='tight')
plt.show()
```

```python
# Group the data by the 'category' column and count the number of papers for each category
category_counts = df1.groupby('category')['id'].count().sort_values(ascending=False)

# Take the top n categories with the most papers
top_n_categories = category_counts.head(15)

# Plot the results using a bar chart
sns.barplot(y=top_n_categories.index, x=top_n_categories.values, orient = 'h')
plt.xlabel('Number of Papers')
plt.ylabel('Categories')
plt.title("(1986 - 2022)", fontsize=10)
plt.suptitle("Trending Research Categories on arXiv", fontsize=14)
plt.xticks(rotation=90)
plt.savefig('Trending Research Categories on arXiv (1986-2022).jpg',dpi=200,bbox_inches='tight')
plt.show()
```

"hep-ph" is a category that stands for High Energy Physics - Phenomenology. 
It's a subfield of High Energy Physics (HEP) that deals with the theoretical study and interpretation of experimental data from particle physics experiments.

Phenomenology, in this context, refers to the study of the observable phenomena in particle physics, such as the properties and interactions of subatomic particles, the behavior of matter and energy at extremely high energies and densities, and the search for new particles and phenomena beyond the current understanding of the standard model of particle physics.

```python
total_categories = df1['category'].nunique()
print(f'Total Number of Categories are: {total_categories}')
```

```python
category_counts = df1['category'].value_counts().sort_values(ascending=False)
category_counts = category_counts.to_frame().reset_index()
category_counts.columns = ['category', 'count']
print(category_counts)
```

```python


category_counts = df1['category'].value_counts().sort_values(ascending=False)
category_counts = category_counts.to_frame().reset_index()
category_counts.columns = ['category', 'count']

# filter the categories with count > 40000
category_counts = category_counts.query("count > 30000")

# merge the filtered categories with original dataframe
df1 = df1.merge(category_counts, on='category')

#group by year and category
df1 = df1.groupby(['year','category']).size().reset_index(name='count')

# plot the data
sns.lineplot(x='year',y='count',hue='category',data=df1)
plt.xlabel('Year')
plt.ylabel('Number of Papers')
plt.title("(1986 - 2022)", fontsize=10)
plt.suptitle("Evolution of Top Research Fields on arXiv", fontsize=14)
plt.savefig('Evolution of Top Research Fields on arXiv (1986-2022).jpg',dpi=200,bbox_inches='tight')
plt.show()
```
